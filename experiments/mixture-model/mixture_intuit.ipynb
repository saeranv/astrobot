```python
import numpy as np 
import pandas as pd
import scipy.stats as stats 
import matplotlib.pyplot as plt 
```

```python
def make_data(N):
    rand = np.random.RandomState(seed=1)
    x = rand.randn(N)
    x[int(0.3 * N):] += 5
    return x 

x = make_data(1000)

plt.hist(x, bins=30);
```

```python
x_d = np.linspace(-4, 8, 1000)
x_ = np.sort(x[::40])  # 1000/50 = 20
_, ax = plt.subplots(1, 5, figsize=(10, 2), sharey=True)

for j in range(5):
    j_ = (j + 1) * 5
    for i in range(j_-1):
        x_bnd = x_[j_-1]
        density = sum(stats.norm(xi).pdf(x_d) for xi in x_[:i+1])   
        ax[j].fill_between(
            x_d, density, alpha=0.5, facecolor='lightblue')
        ax[j].plot(x_d, density, c="blue", alpha=0.5);
        ax[j].plot(np.ones(10)*x_bnd, np.arange(10), c="red")
    ax[j].set_title(f"{j_}%, x:0-{x_[j_-1].round(2)}")
```
