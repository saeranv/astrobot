```python
import numpy as np
from pprint import pprint as pp 
import matplotlib.pyplot as plt
```

```python
# Set parameters
mu = 2
sd = 0.5
x = np.arange(mu-4, mu+4, 0.05)
```

```python
# Make a quadratic
y = x**2
plt.plot(x, y)
plt.grid()
```

```python
# Move quadratic
y = (x - mu)**2
plt.plot(x, y)
plt.grid()
```

```python
# Squish by sigma (note change in y)
y = (x - mu)**2 / sd**2
plt.plot(x, y)
plt.grid()
```

```python
# Getting an intuition for e
pp(np.exp(0))       # 1.0
pp(np.exp(-np.inf)) # 0.0
```

```python
# Based on properties of e: 
# max is when x = mu, min when x is inf
# Thus you get exp curve that maxes at mu
# flattens out based on sd, and flips and 
# does the same in the opposite direction.
y = np.exp(
    -0.5 * ((x - mu)**2) / sd**2
)
plt.plot(x, y)
plt.grid()
```
