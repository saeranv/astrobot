```python
import numpy as np 
from pprint import pprint 
import matplotlib.pyplot as plt 

def pp(x, *args): pprint(x) if not args else (pprint(x), pp(*args))
```

```python
theta = np.arange(360) * np.pi / 180.0
x, y = np.cos(theta), np.sin(theta) 
r = 1 / np.max(np.abs(np.vstack((x, y))), axis=0)

_, ax = plt.subplots(); ax.axis('equal');
for a in np.arange(0, 1.2, 0.2):
    r_ = a + ((1 - a) * r) 
    ax.plot(r_ * x, r_ * y, c='red')

```

```python
N = 20
z = np.linspace(-1, 1, N)
# to get t associated with interval: 
# x = cos(t); acos(x) = t
yz = np.sin(np.arccos(z))

_, ax = plt.subplots(); ax.axis('equal')
ax.plot(x, y, c='r')
for i in range(N):
    ax.plot([0, yz[i]], [z[i], z[i]], c='r')
```
