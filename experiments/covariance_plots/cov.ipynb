```python
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels as sm 
from scipy import stats 
import seaborn as sn 
```

```python
# Covariance matrix
sigma = 0.8
cov = np.array(
    [[1.0, sigma],
     [sigma, 1.0]])
mu = [0, 0]

gauss = stats.multivariate_normal(mu, cov)
q = gauss.rvs(size=1000)

_, ax = plt.subplots(1, 1)
ax.scatter(q[:, 0], q[:, 1], s=2, c='r');
```
