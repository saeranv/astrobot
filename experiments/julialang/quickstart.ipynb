```julia
x = range(0, stop=2*pi, length=1000) 
y = sin.(x)
f, a = plt.subplots(1,1, figsize=(3, 3))
a.plot(x, y, color="red", linewidth=2.0)
a.grid(true)
title("sinusoid")
```

```julia
using StaticArrays 
using LinearAlgebra 
using Statistics

# function 
fx(a, b) = a + sum(b) 
fx(1, 2)

# piping 
ones(3) |> (y -> fx(2, y))

# pretty prints 
a = ones(4)

@show length(a)
@show var(a)
```
